```python
import matplotlib
import matplotlib.pyplot as plt
import pandas
import os
import numpy as np

font = {'size'   : 18}

matplotlib.rc('font', **font)
```

# data for gossiping
##### COMAND: java Constructor GOSSIP 9 500 100 100

```python
samples = []

for i in range(0,100):
    if os.path.isfile("gossip_lob/data_" + str(i) + ".csv"):
        samples.append(pandas.read_csv("gossip_lob/data_" + str(i) + ".csv"))
```

# Ploting for gossiping
### Ploting evolution of the coverage of the network over time

```python
plt.figure(figsize=(20,10))
for sample in samples:
    plt.plot(sample['Communication timer'], sample[' number of nodes spread'])
    
plt.ylabel('Number of nodes covered', size = 22)
plt.xlabel('Time (ms)', size = 22)
plt.title('Evolution of the coverage of the network', size = 22)

plt.xlim(0, 19000)
    
plt.show()
```

### Ploting the time needed for the network to be complete

```python
saver_for_histogram = []
plt.figure(figsize=(20,10))

for sample in samples:
    saver_for_histogram.append(max(sample[' latest time of propagation']))

plt.hist(saver_for_histogram, bins = 5, edgecolor='white', facecolor='g', alpha=0.75)

plt.xlabel('Time taken to propagate to all nodes')
plt.ylabel('Number of ocorrences')
plt.title('Time for the message to get to all nodes (ms)')

plt.show()
```

### Ploting the average number of comunication messages needed for the network to be complete

```python
saver_for_histogram = []
plt.figure(figsize=(20,10))
for sample in samples:
    saver_for_histogram.append(max(sample[' comunication counter']))

plt.xlabel('Number of messages traded in a iteration')
plt.ylabel('Number of ocorrences')
plt.title('Number of comunications needed for the network to be complete')
    
plt.hist(saver_for_histogram, bins = [995.5,996.5,997.5,998.5,999.5,1000.5,1001.5], facecolor='b', alpha=1)
plt.xlim(994, 1001)
plt.show()
```

### Comparison between number of positive and negative acknowledges

```python
pos_ack = []
neg_ack = []

plt.figure(figsize=(20,10))
for sample in samples:
    pos_ack.append(max(sample[' number of positive ack']))
    neg_ack.append(max(sample[' number of negative ack']))

ind = np.arange(50)
width = 0.4

p1 = plt.bar(ind, tuple(pos_ack[:50]), width)
p2 = plt.bar(ind, tuple(neg_ack[:50]), width, bottom=tuple(pos_ack[:50]))

plt.yticks(np.arange(0, 1000, 100))

plt.show()
```
